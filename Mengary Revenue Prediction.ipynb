```python
dummies_depart = pd.get_dummies(df['class'])
df = df.join(dummies_depart)
df.drop(columns=['class','kariox'], axis=1, inplace=True)
df.head()
```

```python
dummies_depart = pd.get_dummies(df['segment'])
df = df.join(dummies_depart)
df.drop(columns = ['segment','Consumer'], axis=1, inplace=True)
df.head()
```

```python
dummies_depart = pd.get_dummies(df['sub-class'])
df = df.join(dummies_depart)
df.drop(columns = ['sub-class','phone'], axis=1, inplace=True)
df.head()
```

```python
dummies_depart = pd.get_dummies(df['delivery type'])
df = df.join(dummies_depart)
df.drop(columns = ['delivery type','Standard Class'], axis=1, inplace=True)
df.head()
```

# Gradient Boosting Model

```python
#Importing necessary packages
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
np.random.seed(3)
import warnings
warnings.filterwarnings('ignore')
```

```python
data_X = df.loc[:, df.columns != "profit"]
data_Y = df[["profit"]]
X_train, X_test ,y_train, y_test = train_test_split(data_X, data_Y, test_size=0.15)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# B = X_train.join(y_train)
# B
```

```python
# # Outlier Removal using the IQR Method (Not used since the model performance is affected later)
# # Income    
# Q1 = B.profit.quantile(0.25)
# Q3 = B.profit.quantile(0.75)
# IQR = Q3 - Q1

# upper_limit = Q3 + 4*IQR

# upper_limit
# B = B[(B.profit<upper_limit)]
```

```python
# X_train = B.loc[:, B.columns != "profit"]
# y_train = B[["profit"]]
```

```python
gradientregressor = GradientBoostingRegressor() 

model = gradientregressor.fit(X_train,y_train)

y_pred = model.predict(X_test)
r2_score(y_pred,y_test)
print(r2_score(y_test,y_pred))
print(mean_squared_error(y_train,model.predict(X_train)))
print(mean_squared_error(y_test,y_pred))
```
